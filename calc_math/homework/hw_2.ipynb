# 1 II.7.26
Доказать, что $$\mu(A·B) ≤ \mu(A)\cdot \mu(B)$$ для любой из норм матриц, согласованных с нормами векторов, и любых квадратных матриц (A, B — квадратные матрицы).


### Доказательство

Пусть $\|\cdot\|$ — норма вектора, а $\mu(\cdot)$ — согласованная норма матрицы:

$$
\mu(A)=\sup_{x\ne 0}\frac{\|Ax\|}{\|x\|}.
$$
Тогда для любых квадратных (A,B) и любого $x\neq0$:
$$
\frac{\|ABx\|}{\|x\|} \le \frac{\|A(Bx)\|}{\|x\|} \le \mu(A) \frac{\|Bx\|}{\|x\|}
\le \mu(A)\mu(B).
$$
Беря супремум по $x\neq0$, получаем:
$$
\ \mu(AB)\le \mu(A)\mu(B)\ }.
$$
---
### Проверка
$$
A=\begin{pmatrix}1 & 2\\ 2 & 1\end{pmatrix},\quad
B=\begin{pmatrix}3 & 1\\ 1 & 3\end{pmatrix}, \quad
AB=\begin{pmatrix}5&7\\7&5\end{pmatrix}.
$$



### Норма $\|\cdot\|_1$ (по столбцам)
$$
\|A\|_1=\max\{\,|1|+|2|,\;|2|+|1|\,\}=3, \qquad
\|B\|_1=\max\{\,|3|+|1|,\;|1|+|3|\,\}=4,
$$
$$
\|AB\|_1=\max\{\,|5|+|7|,\;|7|+|5|\,\}=12.
$$
**Тогда:** 
$$
\|AB\|_1=12=\|A\|_1\,\|B\|_1=3\cdot 4.
$$



### Норма $\|\cdot\|_\infty$ (по строкам)
$$
\|A\|_\infty=\max\{\,|1|+|2|,\;|2|+|1|\,\}=3,\qquad
\|B\|_\infty=\max\{\,|3|+|1|,\;|1|+|3|\,\}=4,
$$
$$
\|AB\|_\infty=\max\{\,|5|+|7|,\;|7|+|5|\,\}=12.
$$
**Тогда:**
$$
\|AB\|_\infty=12=\|A\|_\infty\,\|B\|_\infty=3\cdot 4.
$$



### Норма $\|\cdot\|_2$ 
Матрицы $A,B,AB$ симметричны, поэтому $\|M\|_2=\max_i|\lambda_i(M)|$

**Для $A$:**
$$
\det\!\begin{pmatrix}1-\lambda & 2\\ 2 & 1-\lambda\end{pmatrix}
=(1-\lambda)^2-4=\lambda^2-2\lambda-3=0
\;\Rightarrow\;
\lambda\in\{3,-1\},
$$
$$
\Rightarrow\ \|A\|_2=\max\{|3|,|{-1}|\}=3.
$$

**Для $B$:**
$$
\det\!\begin{pmatrix}3-\lambda & 1\\ 1 & 3-\lambda\end{pmatrix}
=(3-\lambda)^2-1=\lambda^2-6\lambda+8=0
\;\Rightarrow\;
\lambda\in\{4,2\},
$$
$$
\Rightarrow\ \|B\|_2=\max\{|4|,|2|\}=4.
$$

**Для $AB$:**
$$
\det\!\begin{pmatrix}5-\lambda & 7\\ 7 & 5-\lambda\end{pmatrix}
=(5-\lambda)^2-49=\lambda^2-10\lambda-24=0
\;\Rightarrow\;
\lambda\in\{12,-2\},
$$
$$
\Rightarrow\ \|AB\|_2=\max\{|12|,|{-2}|\}=12.
$$

**Тогда:**
$$
\|AB\|_2=12=\|A\|_2\,\|B\|_2=3\cdot 4.
$$


# 2 II.9.2г

Дана система линейных алгебраических уравнений $Ax=b$

Оценить максимально точно относительную погрешность $\dfrac{\|\Delta x\|}{\|x\|}$ в
заданной норме. 

Найти вектор ошибки  $\Delta b$, на котором эта оценка достигается. При каком  $\Delta b$ относительная ошибка $\dfrac{\|\Delta x\|}{\|x\|}$ будет минимальной?

г) $A = \begin{pmatrix}65 & 72\\ 72 & 82\end{pmatrix}$, $\; b = \begin{pmatrix}137\\154\end{pmatrix}; \dfrac{\|\Delta b\|_\infty}{\|b\|_\infty} = 0.01; \|\mathbf x\|_3=\sqrt{(\mathbf x,\mathbf x)}$

### 1) При каком $\Delta b$ относительная ошибка максимальна?
$$
\Delta x=A^{-1}\Delta b,\qquad
\frac{\|\Delta x\|_3}{\|x\|_3}
=\frac{\|A^{-1}\Delta b\|_3}{\|x\|_3}
\le \frac{\|A^{-1}\|_3\,\|\Delta b\|_3}{\|x\|_3}.
$$

Сначала найдём исходное решение:
$$
A\binom{1}{1}=\binom{65+72}{72+82}=\binom{137}{154}=b
\;\Rightarrow\;
x=\binom{1}{1},\quad \|x\|_3=\sqrt2.
$$

Норма правой части:
$$
\|b\|_3=\sqrt{137^2+154^2}=\sqrt{42485}\approx 206.11,
\quad\Rightarrow\quad
\|\Delta b\|_3=0.01\,\|b\|_3.
$$

Так как $A$ симметрична, её собственные значения равны сингулярным:
$$
\det(A-\lambda I)=(65-\lambda)(82-\lambda)-72^2=0
\ \Rightarrow\
\lambda_{\max}=146,\ \lambda_{\min}=1.
$$
Следовательно
$$
\|A^{-1}\|_3=\frac{1}{\lambda_{\min}}=1.
$$

Итого получаем точную (достижимую) верхнюю оценку:
$$

\frac{\|\Delta x\|_3}{\|x\|_3}
\le \|A^{-1}\|_3\,\frac{\|\Delta b\|_3}{\|x\|_3}
=\frac{0.01\,\|b\|_3}{\sqrt2}
=\frac{1}{100}\sqrt{\frac{42485}{2}}
\approx 1.458
$$

В евклидовой максимум $\|A^{-1}\Delta b\|_3$ при фиксированной $\|\Delta b\|_3$ достигается,
когда $\Delta b$ направлен по правому сингулярному вектору $A^{-1}$ для $\sigma_{\max}(A^{-1})=\|A^{-1}\|_3=1$,
то есть по собственному вектору $A$, соответствующему $\lambda_{\min}=1$

Решим $(A-I)v=0$:
$$
\begin{pmatrix}64&72\\72&81\end{pmatrix}\binom{v_1}{v_2}=0
\ \Rightarrow\ v_2=-\frac{64}{72}v_1=-\frac{8}{9}v_1.
$$
Нормированный собственный вектор:
$$
v_{\min}=\frac{1}{\sqrt{145}}\binom{9}{-8}.
$$
При $\varepsilon=0.01$ имеем
$$

\Delta b_{\max}=\varepsilon\,\|b\|_3\,v_{\min}
=\frac{0.01\,\sqrt{42485}}{\sqrt{145}}\binom{9}{-8}
\approx 0.1711\binom{9}{-8}
=\binom{1.540}{-1.369}
$$
Так как $A^{-1}v_{\min}=1\cdot v_{\min}$,
$$
\Delta x_{\max}=A^{-1}\Delta b_{\max}=\Delta b_{\max},\qquad
\frac{\|\Delta x_{\max}\|_3}{\|x\|_3}
=\frac{0.01\,\|b\|_3}{\sqrt2}\approx 1.458.
$$

---

### 2) При каком $\Delta b$ относительная ошибка минимальна?
Минимум $\|A^{-1}\Delta b\|_3$ при фиксированной $\|\Delta b\|_3$ достигается,
когда $\Delta b$ направлен по собственному вектору $A$, соответствующему $\lambda_{\max}=146$
(потому что тогда $A^{-1}$ масштабирует на $1/146$).

Из $(A-146I)u=0$ получаем $u_2=\dfrac{81}{72}u_1=\dfrac{9}{8}u_1$, и
$$
v_{\max}=\frac{1}{\sqrt{145}}\binom{8}{9}.
$$
Берём
$$

\Delta b_{\min}=\varepsilon\,\|b\|_3\,v_{\max}
=\frac{0.01\,\sqrt{42485}}{\sqrt{145}}\binom{8}{9}
\approx 0.1711\binom{8}{9}
=\binom{1.369}{1.540}
$$
Тогда $A^{-1}v_{\max}=\dfrac{1}{146}v_{\max}$ и
$$

\frac{\|\Delta x_{\min}\|_3}{\|x\|_3}
=\frac{1}{146}\cdot\frac{0.01\,\|b\|_3}{\sqrt2}
=\frac{1}{14600}\sqrt{\frac{42485}{2}}
\approx 0.00999
$$



# 3 II.7.37(б)

Найти область сходимости метода Якоби и метода Зейделя для систем с матрицами вида:

б) $$\mathbf{A} = \begin{pmatrix} \alpha & 0 & \beta \\ 0 & \alpha & 0 \\ \beta & 0 & \alpha \end{pmatrix}$$

Матрицу $ \mathbf{A} $ можно представить в виде суммы:

$$
\mathbf{A} = \mathbf{D} + \mathbf{L} + \mathbf{U},
$$

где:

- $ \mathbf{D} $ — диагональная часть:

  $$
  \mathbf{D} = \begin{pmatrix}
  \alpha & 0 & 0 \\
  0 & \alpha & 0 \\
  0 & 0 & \alpha
  \end{pmatrix}.
  $$

- $ \mathbf{L} $ — нижнетреугольная часть без диагонали:

  $$
  \mathbf{L} = \begin{pmatrix}
  0 & 0 & 0 \\
  0 & 0 & 0 \\
  \beta & 0 & 0
  \end{pmatrix}.
  $$

- $ \mathbf{U} $ — верхнетреугольная часть без диагонали:

  $$
  \mathbf{U} = \begin{pmatrix}
  0 & 0 & \beta \\
  0 & 0 & 0 \\
  0 & 0 & 0
  \end{pmatrix}.
  $$

## 1) Якоби
$$
\mathbf{B}_{\text{Якоби}} = -\mathbf{D}^{-1} (\mathbf{L} + \mathbf{U}).
$$

$$
\mathbf{B}_{\text{Якоби}} = -\dfrac{1}{\alpha}
\begin{pmatrix}
  0 & 0 & \beta \\
  0 & 0 & 0 \\
  \beta & 0 & 0
  \end{pmatrix}
$$

$$ \det\begin{pmatrix} 
\alpha \lambda & 0 & \beta \\
0 & \alpha \lambda & 0 \\
\beta & 0 & \alpha \lambda
\end{pmatrix} = 0 = (-\lambda \alpha)((\lambda \alpha)^2-\beta^2)
$$

Следовательно,
$$
\lambda=\left\{0,\ \pm\frac{\beta}{\alpha}\right\},\qquad
\rho(B_{\text{Якоби}})=\frac{|\beta|}{|\alpha|} \leq 1.
$$

Ответ: $|\alpha| \leq |\beta|$




## 2) Зейделя
$$
B_{\text{Зейделя}}=-(D+L)^{-1}U,\qquad
D+L=\begin{pmatrix}
\alpha&0&0\\
0&\alpha&0\\
\beta&0&\alpha
\end{pmatrix}.
$$

Так как $D+L$ нижнетреугольная,
$$
(D+L)^{-1}=
\begin{pmatrix}
\frac1\alpha&0&0\\[3pt]
0&\frac1\alpha&0\\[3pt]
-\frac{\beta}{\alpha^2}&0&\frac1\alpha
\end{pmatrix}.
$$

Тогда
$$
B_{\text{Зейделя}}
=
-\begin{pmatrix}
\frac1\alpha&0&0\\
0&\frac1\alpha&0\\
-\frac{\beta}{\alpha^2}&0&\frac1\alpha
\end{pmatrix}
\begin{pmatrix}
0&0&\beta\\
0&0&0\\
0&0&0
\end{pmatrix}
=
\begin{pmatrix}
0&0&-\frac{\beta}{\alpha}\\[4pt]
0&0&0\\[4pt]
0&0&\frac{\beta^2}{\alpha^2}
\end{pmatrix}.
$$

Матрица верхнетреугольная, значит её собственные значения — элементы диагонали:
$$
\lambda(B_{\text{Зейделя}})=\left\{0,\,0,\,\frac{\beta^2}{\alpha^2}\right\},\qquad
\rho(B_{\text{Зейделя}})=\frac{\beta^2}{\alpha^2}\leq 1
$$

**Сходимость Зейделя:**
$$
|\alpha|\leq |\beta|
$$



# 4  1*д

Найдем функцию $\psi(x)$, удовлетворяющую
$$
-\psi''(x) = \sin(2\pi x),\quad x\in(0,1),\qquad \psi(0)=10,\ \psi(1)=20.
$$

Построим равномерную сетку: $x_i = i\cdot h$, $i=0,1,\dots,N$, $h=1/N$

Для внутренних узлов $i=1,\dots,N-1$ используем центральную разность второго порядка:
$$
\psi''(x_i) \approx \frac{\psi_{i+1}-2\psi_i+\psi_{i-1}}{h^2}.
$$

Подставляя в уравнение, получаем СЛАУ
$$
\frac{-\psi_{i+1}+2\psi_i-\psi_{i-1}}{h^2} = \sin(2\pi x_i),\qquad i=1,\dots,N-1.
$$

Обозначая вектор неизвестных значений во внутренних узлах через $u\in\mathbb{R}^{N-1}$,
получаем матричную форму $A u = b$, где
$$
A = \operatorname{tridiag}(-1,2,-1),\qquad
b_i = h^2\sin(2\pi x_i),\ b_1=\psi(0),\ b_{N-1}=\psi(1).
$$

Матрица $A$ симметрична и положительно определена (SPD).


1. Метод наискорейшего спуска (МНС)

Решение СЛАУ $Au=f$ для SPD-матрицы эквивалентно минимизации
$$
\Phi(u)=\tfrac12 u^\top A u - f^\top u.
$$
Невязка $r=f-Au$ связана с градиентом: $\nabla\Phi(u)=Au-f=-r$
Итерация МНС:
$$
u_{k+1}=u_k+\tau_k r_k,\qquad r_k=f-Au_k.
$$
Оптимальный шаг $\tau_k$ получается из одномерной минимизации $\Phi(u_k+\tau r_k)$:
$$
 \tau_k=\dfrac{(r_k,r_k)}{(r_k, A r_k)}
$$

2. Степенной метод и оценка числа обусловленности

Степенной метод применим для оценки $\lambda_{\max}(A)$ Чтобы оценить $\lambda_{\min}(A)$,
используем степенной метод к $A^{-1}$: на каждой итерации решаем $Az=x$
Тогда максимальное собственное значение для $A^{-1}$ равно $1/\lambda_{\min}(A)$
Число обусловленности в евклидовой норме:
$$
\mu_2(A)=\frac{\lambda_{\max}(A)}{\lambda_{\min}(A)}
$$

3. Критерий остановки

Оценка $\|u-u^*\|_2\le \|A^{-1}\|_2\,\|r\|_2$ даёт
$$
\frac{\|u-u^*\|_2}{\|u^*\|_2} \;\lesssim\; \mu_2(A)\,\frac{\|r\|_2}{\|f\|_2}.
$$
Поэтому для гарантии относительной точности $\varepsilon$ достаточно потребовать
$$
\frac{\|r_k\|_2}{\|f\|_2} \le \frac{\varepsilon}{\mu_2(A)}.
$$

4. Оценка скорости сходимости и поведение шага $\tau_k$

Невязка $r_k=b-Au_k$ на каждом шаге умножается на $A$ (в формуле $\tau_k=(r_k,r_k)/(r_k,Ar_k)$), что усиливает высокочастотные компоненты (с большими собственными числами $\lambda_i$). Поэтому
$$
\tau_k=\frac{(r_k,r_k)}{(r_k,Ar_k)} \ \longrightarrow\ \frac{1}{\lambda_{\max}}.
$$









```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la

def build_system(N, psi0=10.0, psi1=20.0):
    """Собирает A, b"""
    h = 1.0/N
    x = np.linspace(0.0, 1.0, N+1)
    xi = x[1:-1]  # внутренние узлы
    main = 2*np.ones(N-1)
    off  = -1*np.ones(N-2)
    A = np.diag(main) + np.diag(off,1) + np.diag(off,-1)
    b = (h**2) * np.sin(2*np.pi*xi)
    b[0]  += psi0
    b[-1] += psi1
    return A, b, x, xi, h

def steepest_descent(A, b, tol_res_rel, maxit=500000):
    """МНС с оптимальным шагом. Останавливается при ||r||/||b|| <= tol_res_rel."""
    x = np.zeros_like(b)
    bnorm = np.linalg.norm(b)
    r = b - A @ x
    res_rel_hist = [np.linalg.norm(r)/bnorm]
    tau_hist = []
    for k in range(maxit):
        Ar = A @ r
        rr = float(r @ r)
        rAr = float(r @ Ar)
        tau = rr / rAr if rAr > 0 else 1.0
        x = x + tau * r
        r = r - tau * Ar  # обновление невязки
        tau_hist.append(tau)
        res_rel = np.linalg.norm(r)/bnorm
        res_rel_hist.append(res_rel)
        if res_rel <= tol_res_rel:
            break
    return x, {"res_rel": res_rel_hist, "tau": tau_hist, "iters": len(tau_hist)}

def power_method(A, n_it=400, x0=None):
    """Степенной метод для оценки lambda_max(A)"""
    n = A.shape[0]
    rng = np.random.default_rng(0)
    x = rng.standard_normal(n) if x0 is None else x0.copy()
    x /= np.linalg.norm(x)
    lam = None
    for _ in range(n_it):
        y = A @ x
        x = y / np.linalg.norm(y)
        lam = float(x @ (A @ x))
    return lam

def power_method_inv(A, solver, n_it=400, x0=None):
    """Степенной метод для A^{-1}: возвращает оценку mu ~ 1/lambda_min(A)."""
    n = A.shape[0]
    rng = np.random.default_rng(1)
    x = rng.standard_normal(n) if x0 is None else x0.copy()
    x /= np.linalg.norm(x)
    mu = None
    for _ in range(n_it):
        z = solver(x)      # z = A^{-1} x
        x = z / np.linalg.norm(z)
        mu = float(x @ solver(x))  # x^T A^{-1} x
    return mu

def exact_solution(x):
    """Точное решение: psi(x) = sin(2*pi*x)/(4*pi^2) + 10*x + 10."""
    return (np.sin(2*np.pi*x)/(4*np.pi**2)) + 10.0*x + 10.0
```

```python
N = 100
eps = 1e-5

A, b, x_all, xi, h = build_system(N)

lu, piv = la.lu_factor(A)
solveA = lambda rhs: la.lu_solve((lu, piv), rhs)


lam_max = power_method(A, n_it=400)
mu_max  = power_method_inv(A, solver=solveA, n_it=400)
lam_min = 1.0 / mu_max
kappa   = lam_max / lam_min
print(f"lambda_max ~ {lam_max:.6f},  lambda_min ~ {lam_min:.6f},  kappa(A) ~ {kappa:.2f}")


tol_res_rel = eps / kappa
print(f"Порог относительной невязки: eps/kappa = {tol_res_rel:.3e}")


u_int, hist = steepest_descent(A, b, tol_res_rel=tol_res_rel, maxit=200000)
print(f"Итераций МНС: {hist['iters']}")
print(f"Достигнутая относительная невязка: {hist['res_rel'][-1]:.3e}")


psi = np.zeros(N+1)
psi[0], psi[-1] = 10.0, 20.0
psi[1:-1] = u_int


psi_ex = exact_solution(x_all)
rel_err = np.linalg.norm(psi-psi_ex)/np.linalg.norm(psi_ex)
print(f"Относительная ошибка к точному: {rel_err:.3e}")


plt.figure(figsize=(6,4))
plt.plot(x_all, psi_ex, label='точное')
plt.plot(x_all, psi, '--', label='МНС (разности)')
plt.xlabel('x'); plt.ylabel('psi(x)'); plt.title('Решение задачи')
plt.legend(); plt.grid(alpha=0.3); plt.show()

# 2) Сходимость по невязке
plt.figure(figsize=(6,4))
plt.semilogy(hist['res_rel'])
plt.xlabel('k'); plt.ylabel(r'$||r_k||/||b||$'); plt.title('Сходимость МНС')
plt.grid(alpha=0.3); plt.show()

# 3) Зависимость tau_k от k
j = np.arange(1, N)
lam_all = 4*np.sin(j*np.pi/(2*N))**2 
tau_lo = 1.0 / lam_all.max()
tau_hi = 1.0 / lam_all.min()

plt.figure(figsize=(6,4))
plt.plot(hist['tau'], label=r'$\tau_k$')
plt.axhline(tau_lo, ls='--', label=r'$1/\lambda_{\max}$')
plt.axhline(tau_hi, ls=':', label=r'$1/\lambda_{\min}$')
ax = plt.gca()
ax.ticklabel_format(style='plain', useOffset=False, axis='y')
plt.xlabel('k'); plt.ylabel('tau_k'); plt.title('Зависимость шага $\\tau_k$ от итерации')
plt.legend(); plt.grid(alpha=0.3); plt.show()

# 4) Экспериментальная скорость: q_k = res[k+1]/res[k]
res = np.array(hist['res_rel'])
q = res[1:] / res[:-1]

m = max(3, int(0.3*len(q)))           # «хвост» = последние 30%
q_tail = q[-m:]
q_emp = float(np.median(q_tail))
print(f'Эмпирическая скорость на хвосте (медиана последних {m} значений): {q_emp:.6f}')

# Теоретическая оценка для МНС
q_theor = ((kappa - 1.0) / (kappa + 1.0))**2

plt.figure(figsize=(6,4))
plt.plot(q, label=r'$q_k$')
plt.axhline(q_emp, ls='--', label=f'медиана хвоста ≈ {q_emp:.4f}')
plt.axhline(q_theor, ls=':', label=f'оценка теории ≈ {q_theor:.4f}')
plt.xlabel('k'); plt.ylabel(r"$q_k = \|r_{k+1}\|/\|r_k\|$")
plt.title('Экспериментальная скорость сходимости')
plt.legend(); plt.grid(alpha=0.3); plt.show()
```
